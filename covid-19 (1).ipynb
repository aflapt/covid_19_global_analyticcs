**merging common columns diffrent dataset**

combine common columns of df_country and df_covid 

```python
country_columns=set(df_country.columns)
covid_columns=set(df_covid.columns)
common_columns=(country_columns & covid_columns)
common_columns
```

```python
column_common=["Active","Confirmed","Country/Region","Deaths","Recovered","WHO Region"]
merge_covi_country=pd.merge(df_country,df_covid,on=column_common)
merge_covi_country.head(5)
```

**checking outliers**

```python
# Ploting box plot for Confirmed

sns.set(style="darkgrid")  
cases = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# Create a box plot for every cases
plt.figure(figsize=(12, 6))  
sns.boxplot(data=merge_covi_country[cases])


# Add labels and title (optional)
plt.xlabel(f'{cases} cases')
plt.title('Box Plot of Corona distribution')

plt.show()
```
